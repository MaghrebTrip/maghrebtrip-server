```python
import pandas as pd
```

```python
monuments = pd.read_csv('attractions/monuments.csv')
new_monuments = pd.read_csv('attractions/new_monuments.csv', sep=";")
restaurants = pd.read_csv('attractions/restaurants.csv')
new_restaurants = pd.read_csv('attractions/new_restaurants.csv', sep=";")
hotels = pd.read_csv('attractions/hotels.csv')
new_hotels = pd.read_csv('attractions/new_hotels.csv', sep=";")
```

```python
new_monuments
```

```python
# replace rabat monuments in monunemts.csv with the new ones in new_monuments.csv
monuments_city_id = 1
# drop the old monuments
monuments = monuments[monuments['city_id'] != monuments_city_id]
```

```python
monuments[monuments['city_id'] == monuments_city_id]
```

```python
for index, row in new_monuments.iterrows():
    monuments = monuments._append(row)
```

```python
monuments[monuments['city_id'] == monuments_city_id]
```

```python
# same for restaurants
restaurants_city_id = 1

restaurants = restaurants[restaurants['city_id'] != restaurants_city_id]

restaurants[restaurants['city_id'] == restaurants_city_id]
```

```python
for index, row in new_restaurants.iterrows():
    restaurants = restaurants._append(row)

restaurants[restaurants['city_id'] == restaurants_city_id]
```

```python
# same for hotels
hotels_city_id = 1

hotels = hotels[hotels['city_id'] != hotels_city_id]

hotels[hotels['city_id'] == hotels_city_id]
```

```python
for index, row in new_hotels.iterrows():
    hotels = hotels._append(row)

hotels[hotels['city_id'] == hotels_city_id]
```

```python
# sort by city_id then reset index
monuments = monuments.sort_values(by=['city_id']).reset_index(drop=True)
restaurants = restaurants.sort_values(by=['city_id']).reset_index(drop=True)
hotels = hotels.sort_values(by=['city_id']).reset_index(drop=True)
```

```python
# save the new files
monuments.to_csv('attractions/monuments.csv', index=False)
restaurants.to_csv('attractions/restaurants.csv', index=False)
hotels.to_csv('attractions/hotels.csv', index=False)
```
